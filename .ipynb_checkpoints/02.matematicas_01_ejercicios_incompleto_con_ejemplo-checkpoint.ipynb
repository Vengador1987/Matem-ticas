### Ejercicio 2:
- Muestra 20 puntos aleatorios entre los dos punto de la recta anterior como se muestra en la gr√°fica

```python
P1 = (-2, -2)
P2 = (2, 5)
```

```python
x, y = symbols("x, y")

ecuacion = ((x - P1[0]) / (P2[0] - P1[0])) - ((y - P1[1]) / (P2[1] - P1[1]))

ecuacion_lineal = solve(ecuacion, y)[0]

ecuacion_lineal
```

```python
#from random import randint

#num_aleatorios = [randint(0, 100) for i in range(20)]
#num_aleatorios
```

```python
num_aleatorios = np.random.randint(0, 100, 20)
num_aleatorios
```

```python
resultados = list()

for i in num_aleatorios:
    
    resultado = ecuacion_lineal.subs(x, i)
    
    resultados.append(resultado.evalf())
    
resultados
```

```python
for x, y in zip(num_aleatorios, resultados):
    plt.plot(x, y, marker = "o", color = "red")
    
plt.plot(num_aleatorios, resultados)
plt.show()
```

### Ejercicio 3:
- Encuentra el punto donde cruzan las siguientes ecuaciones y muestralo en una gr√°fica 

$$ x - 0.3y -3 = 0 $$
$$ 0.5x-y +2 = 0 $$




```python
x, y = symbols("x, y")

eq1 = x - 0.3*y - 3
eq2 = 0.5*x - y + 2

eq1_solve = solve(eq1, y)[0]
eq2_solve = solve(eq2, y)[0]

eq1_solve, eq2_solve
```

```python
px = solve(eq1.subs(y, eq2_solve))[0]
py = eq1_solve.subs(x, px)
px, py
```

```python
X = np.linspace(-10, 10, 100)

y1 = [eq1_solve.subs(x, i) for i in X]
y2 = [eq2_solve.subs(x, i) for i in X]

# Rectas
plt.plot(X, y1, color = "blue")
plt.plot(X, y2, color = "red")

# Punto
plt.plot(px, py, marker = "o", color = "green")

plt.show()
```

```python
def interseccion_rectas(eq1, eq2):
    
    eq1_solve = solve(eq1, y)[0]
    eq2_solve = solve(eq2, y)[0]
    
    px = solve(eq1.subs(y, eq2_solve))[0]
    py = eq1_solve.subs(x, px)

    X = np.linspace(-10, 10, 100)

    y1 = [eq1_solve.subs(x, i) for i in X]
    y2 = [eq2_solve.subs(x, i) for i in X]

    # Rectas
    plt.plot(X, y1, color = "blue")
    plt.plot(X, y2, color = "red")

    # Punto
    plt.plot(px, py, marker = "o", color = "green")

    plt.show()
```

```python
x, y = symbols("x, y")

eq1 = -7*x + 0.8*y - 10
eq2 = -6*x + 8*y + 90

interseccion_rectas(eq1, eq2)
```

### Ejercicio 4:
- Encuentra el resultado de la siguiente ecuaci√≥n y muestralo en una gr√°fica

$$ x^2 -5x -6 = 0$$

```python
x = Symbol("x")

eq = x**2 - 5*x - 6

raices = solve(eq)

raices
```

```python
X = np.linspace(-20, 20, 100)
y = [eq.subs(x, i) for i in X]

# Recta
plt.plot(X, y)

# Puntos
for raiz in raices:
    plt.plot(raiz, 0, marker = "o", color = "red")

plt.show()
```

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calcular y graficar la ecuaci√≥n de una recta
```

```python
P1 = (5, 4)
P2 = (2, 3)
```

```python
x, y = symbols("x, y")

eq = ((x - P1[0]) / (P2[0] - P1[0])) - ((y - P1[1]) / (P2[1] - P1[1]))

ecuacion_recta = solve(eq, y)[0]

ecuacion_recta
```

```python
X = np.linspace(-5, 5, 10)
Y = [ecuacion_recta.subs(x, i) for i in X]

# Punto 1
plt.plot()

# Recta
plt.plot(X, Y)
plt.show()
```

```python
dict_v = {a : b for a, b in zip(Y, X)}
dict_v
```

```python
plt.plot(dict_v.values(), dict_v.keys(), marker = "o")
```

```python
# Interseccion 2 rectas:

x, y = symbols("x, y")

eq1 = 2*x + 5 - y
eq2 = x/2 + y
```

```python
eq1_solve_y = solve(eq1, y)[0]
eq1_solve_y
```

```python
eq2_solve_y = solve(eq2, y)[0]
eq2_solve_y
```

```python
px = solve(eq2.subs(y, eq1_solve_y))[0]
px
```

```python
py = solve(eq1.subs(x, px))[0]
py
```

```python
PI = (px, py)
PI
```

```python
X = np.linspace(-5, 5, 10)

Y1 = [eq1_solve_y.subs(x, i) for i in X]
Y2 = [eq2_solve_y.subs(x, i) for i in X]

# Recta 1
plt.plot(X, Y1, color = "red")

# Recta 2
plt.plot(X, Y2, color = "blue")

# Punto I
plt.plot(PI[0], PI[1], color = "green", marker = "o")

plt.show()
```
