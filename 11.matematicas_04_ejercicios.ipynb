```python
import numpy as np 
import matplotlib.pyplot as plt

import math 
from sympy import * 
```

### Ejercicio 1:
- Grafica la función del coseno para el rango de -10 a 10.

```python
x = np.linspace(-10, 10, 100)
y = np.cos(x)

plt.plot(x, y, color = "black")

plt.xlim(-10, 10)
plt.ylim(-2, 2)
plt.grid()
plt.show()
```

### Ejercicio 3:
- Calcula la derivada de la siguiente ecuación: $$ sin(x^2)+10 $$

```python
x = Symbol('x') 
fx = sin(x**2) + 10

Derivative(fx, x).doit()
```

Dada la siguiente función: 

$$x^2+10x+1 = 0$$ 

- Genera la grafica de la función, encuentra su derivada y muestra la recta tangente de la derivada en el punto x = -7.5

```python
x = np.linspace(-10, 10, 100)
fx = x**2 + 10*x + 1

plt.plot(x, fx, color = "black")
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = m*x - m*-7.5 + (-17.75)  #  y = mx-mx1+y1 : En este ejemplo, da -17.75 (sustituir x = -7.5 en fx)

plt.plot(x, y, color = "black")
plt.grid()
plt.show()
```

```python
# Para x = -7.5, y = -17.75

m = (2 * -7.5) + 10 # Pendiente
x = np.linspace(-10, 10, 100)

fx = x**2 + 10*x + 1 # Funcion f(x)
rectaTangente = m*x - m*-7.5 + (-17.75) # Ecuacion de la recta (para los valores en x e y)

plt.plot(x, fx) # Funcion f(x)
plt.plot(-7.5, -17.75, marker = "o") #Punto
plt.plot(x, rectaTangente) # Recta Tangente

plt.grid()
plt.show()
```

```python
x = Symbol("x")
           
fx = x**2 + 10*x + 1

px = -7.5
py = fx.subs(x, px)

derivada = Derivative(fx, x).doit()
m = derivada.subs(x, px)

recta_tangente = m*x - m*px + py

print(f"Función : {fx}")
print(f"Punto: ({px}, {py})")
print(f"Derivada: {derivada}")
print(f"Recta Tangente: {recta_tangente}")
```
