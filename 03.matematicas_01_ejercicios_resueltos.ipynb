- Muestra 20 puntos aleatorios entre los dos punto de la recta anterior como se muestra en la gráfica

```python
P1 = (-2, -2)
P2 = (2, 5)
```

```python
# Ecuación de la recta
x, y = symbols("x, y")

ecuacion = ((x - P1[0]) / (P2[0] - P1[0])) - ((y - P1[1]) / (P2[1] - P1[1]))

ecuacion_recta = solve(ecuacion, y)[0]

ecuacion_recta
```

### Ejercicio 3:
- Encuentra el punto donde cruzan las siguientes ecuaciones y muestralo en una gráfica 

$$ x - 0.3y -3 = 0 $$
$$ 0.5x-y +2 = 0 $$




```python
# Ecuaciones 

x, y = symbols("x, y")

eq1 = x - 0.3*y - 3
eq2 = 0.5*x - y + 2

# Ecuaciones en funcion de "x", con la "y" despejada

eq1_solve = solve(eq1, y)[0]
eq2_solve = solve(eq2, y)[0]

eq1_solve, eq2_solve
```

```python
# Calculo del punto de interseccion

px = solve(eq1.subs(y, eq2_solve))[0]
py = eq1_solve.subs(x, px)
px, py
```

```python
X = np.linspace(-10, 10, 100)

y1 = [eq1_solve.subs(x, i) for i in X]
y2 = [eq2_solve.subs(x, i) for i in X]

# Rectas
plt.plot(X, y1, color = "blue")
plt.plot(X, y2, color = "red")

# Punto Interseccion
plt.plot(px, py, marker = "o", color = "green")

plt.show()
```

```python
# Función
def interseccion_rectas(eq1, eq2):
    
    eq1_solve = solve(eq1, y)[0]
    eq2_solve = solve(eq2, y)[0]
    
    px = solve(eq1.subs(y, eq2_solve))[0]
    py = eq1_solve.subs(x, px)

    X = np.linspace(-10, 10, 100)

    y1 = [eq1_solve.subs(x, i) for i in X]
    y2 = [eq2_solve.subs(x, i) for i in X]

    # Rectas
    plt.plot(X, y1, color = "blue")
    plt.plot(X, y2, color = "red")

    # Punto
    plt.plot(px, py, marker = "o", color = "green")

    plt.show()
```

```python
# Ejemplo

x, y = symbols("x, y")

eq1 = -7*x + 0.8*y - 10
eq2 = -6*x + 8*y + 90

interseccion_rectas(eq1, eq2)
```

### Ejercicio 5:
- Define una función que le pida a un usuario que ingrese los coeficientes de una ecuación de segundo grado.
- Los parámetros de entrada serán los coeficientes de la ecuación y la función debe retornar los valores **x1** y **x2**, **las raice**s o soluciones de la ecuación, y mostrar la **grafica de la curva** de la ecuación, los puntos **x1** y **x2** y los **ejes X e Y**.
- Si la ecuación no tiene solución en los números reales, la función solo debe mostrar la **gráfica** con los **ejes X e Y**. 

```python
# Función usando la libreria math

def ecuacion_cuadratica_1(a, b, c):
    
    # x, y son np.arrays
    x = np.linspace(-10, 10, 100)
    y = a*x**2 + b*x + c
    
    # Curva
    plt.plot(x, y, color = "black")
    
    # Eje X
    plt.plot(np.linspace(-10, 10, 30), [0 for i in range(30)], color = "black", alpha = 0.5) 

    # Eje Y
    plt.plot([0 for i in range(30)], np.linspace(-1000, 50, 30), color = "black", alpha = 0.5) 
    
    try : 
        x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
        x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)
        
        # Raíz 1
        plt.plot(x1, 0, color = "red", marker = "o")

        # Raíz 2
        plt.plot(x2, 0, color = "red", marker = "o")
        
        plt.show()
        
        return x1, x2
        
    except:
        print("Raíces imaginarias")
        plt.show()
```

```python
# Función usando SymPy

def ecuacion_cuadratica_2(a, b, c):
    
    x = Symbol("x")
    
    eq = a*x**2 + b*x + c
    
    x1, x2 = solve(eq)
    
    X = np.linspace(-10, 10, 100)
    Y = [eq.subs(x, i) for i in X]
    
    # Curva
    plt.plot(X, Y, color = "black")
    
    # Eje X
    plt.plot(np.linspace(-10, 10, 30), [0 for i in range(30)], color = "black", alpha = 0.5) 

    # Eje Y
    plt.plot([0 for i in range(30)], np.linspace(-30, 150, 30), color = "black", alpha = 0.5) 
    
    try:
        # Raíz 1
        plt.plot(x1, 0, color = "red", marker = "o")

        # Raíz 2
        plt.plot(x2, 0, color = "red", marker = "o")
        
        plt.show()
        
        return x1, x2
    
    except:
        plt.show()
```

```python
ecuacion_cuadratica_1(-10, 6, 2.5)
```

```python
ecuacion_cuadratica_2(1, -5, -6)
```

### Ejercicio 10:
- Define una función para encontrar la distancia entre dos puntos para **`n`** dimensiones.

```python
def distancia_euclideana(P1, P2):
    
    total = 0
    
    for x, y in zip(P1, P2):
        total += (y - x)**2
    
    total = np.sqrt(total)
    
    return total
```

```python
def distancia_euclideana_2(P1, P2):
    return np.sqrt(sum([(y - x)**2 for x, y in zip(P1, P2)]))
```

```python
P1 = (-2, -3, -4)
P2 = (3, 4, 2)

distancia_euclideana(P1, P2)
```

### Ejercicio 12:
- Crea una función para la **distancia Manhhatan** y otra para la **distancia Chebyshev**. Encuentra las distancias entre el punto A y los puntos B y C del ejercicio anterior utilizando las dos funciones.

```python
def distancia_manhattan(P1, P2):
    
    total = 0
    
    for x, y in zip(P1, P2):
        total += abs(y - x)
        
    return total

def distancia_chebyshev(P1, P2):
    
    distancias = list()
    
    for x, y in zip(P1, P2):
        distancias.append(abs(y - x))
        
    return max(distancias)
```

```python
print(distancia_manhattan(A, B))
print(distancia_manhattan(A, C))
print(distancia_manhattan(B, C))
```

```python
print(distancia_chebyshev(A, B))
print(distancia_chebyshev(A, C))
print(distancia_chebyshev(B, C))
```
